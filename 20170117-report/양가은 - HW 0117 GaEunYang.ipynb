# Product of Matrix

```python
a= [
    [1,2,5],
    [3,4,7],
    [10,11,9]
]
b= [
    [5,6],
    [7,8],
    [1,2]
]

# a11*b11+a12*b21 , a11*b12+a12*b22
# a21*b11+a22*b21 , a21*b12+a22*b22
```

```python
[[sum([a[i][j]*b[j][k] for j in range(len(a[0]))]) for k in range(len(b[0]))] for i in range(len(a))]
```

```python
ep_url ="http://comic.naver.com/webtoon/detail.nhn?titleId=25455&no=443&weekday=tue"
import requests
html = requests.get(ep_url).text

from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
       
import os 
for idx, img_tag in enumerate(soup.select(".wt_viewer img"), 1):
    img_url = img_tag['src']
    
    # ep_url, img_url
    filename = os.path.basename(img_url)
    filepath = os.path.join("data", "노블레스", filename)  # 웹툰ID, EP ID, 이미지name
    dirpath = os.path.dirname(filepath)
    if not os.path.exists(dirpath):
        os.makedirs(dirpath)
    
    if os.path.exists(filepath):
        print('이미 존재 :', filename)
    else:
        print('다운받을 것이오 :', filename)
        headers = {
            'Referer': ep_url,
            #'User-Agent': '',
        }
        image_data = requests.get(img_url, headers=headers).content
        with open(filepath, 'wb') as f:
            f.write(image_data)
```
