```python
# 에피소드 이름 및 url 크롤링

with open("하루세컷_url.txt", "wt") as f :
    for i in range(10) :
        webtoon_url = "http://comic.naver.com/webtoon/list.nhn?titleId=644180&weekday=sun&page={}".format(i+1)

        html = requests.get(webtoon_url).text

        from bs4 import BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")
        
        for img_tag in soup.select('.title a') :
            title = img_tag.contents
            ep_url = img_tag['href']
            
            f.write(str(title))
            f.write(" > ")
            f.write("comic.naver.com")
            f.write(ep_url)
            f.write('\n')
            
```

```python
# 각 에피소드 대표 이미지 크롤링

import requests
import os

for i in range(10) :
    webtoon_url = "http://comic.naver.com/webtoon/list.nhn?titleId=644180&weekday=sun&page={}".format(i+1)
        
    html = requests.get(webtoon_url).text
        
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')
        
    for img_tag in soup.select('td img') :
            
        try :
            img_url = img_tag['src']
            title = img_tag['title']
            
            filename = title + os.path.basename(img_url)
            filepath = os.path.join('하루세컷_img', filename)
            dirpath = os.path.dirname(filepath)
                
            if not os.path.exists(dirpath) :
                os.makedirs(dirpath)
                
            if os.path.exists(filepath) :
                print('이미 존재 : ', title)
            else :
                print('다운로드 : ', title)
                    
                headers = {
                    'Referer' : webtoon_url
                }
                    
                img_data = requests.get(img_url, headers = headers).content
                    
                with open(filepath, 'wb') as f :
                    f.write(img_data)
            
        except Exception as ex :
            pass
```
